```python
import lightgbm as lgb
import pandas as pd
from sklearn.metrics import mean_squared_error

features_selected =  ["service_type","online_time","1_total_fee","2_total_fee", "3_total_fee", "4_total_fee",
                     "month_traffic", "many_over_bill", "contract_type","contract_time","pay_num","last_month_traffic",
                     "local_trafffic_month","local_caller_time","service1_caller_time","service2_caller_time", "age"]

X = df[features_selected]
y = df["res"]
X_result = df1[features_selected]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

#create dataset for lightgbm
#传进来的是dataframe或者series
lgb_train = lgb.Dataset(X_train,y_train)
lgb_eval = lgb.Dataset(X_test,y_test,reference=lgb_train)

params = {
    'task': 'train',
    'boosting_type': 'gbdt',
    'objective': 'multiclass',
    'num_class': 15,
    'metric': {'rmse'},  #'metric': {'l2', 'auc'},
    'num_leaves':39,
    'learning_rate': 0.1,
    'feature_fraction': 0.9,
    'bagging_fraction': 0.8,
    'bagging_freq': 5,
    'verbose': 0,
    'num_iterations': 100,
    'max_depth': -1
    }

muti_results = lgb.LGBMClassifier(
    boosting_type = 'gbdt',
    num_leaves = 39,
    max_depth = 6,
    learning_rate = 0.02,
    n_estimators = 10000,
    max_bin = 255,
    subsample_for_bin = 200000,
    objective = 'multiclass',
    min_split_gain = 0.0,
    min_child_weight = 0.001,
    min_child_samples = 20,
    subsample = 1.0,
    subsample_freq = 1,
    colsample_bytree = 1.0,
    reg_alpha = 0.0,
    reg_lambda = 0.0,
    random_state = 0,
    n_jobs = 3,
    silent = True,
    early_stopping_rounds=200,
    verbose_eval=100,
    show_stdv=True)
muti_results.fit(X_train,
                 y_train,
                 eval_set=[(X_test, y_test)],
                 early_stopping_rounds = 200
                )

print(muti_results.best_iteration_)
y_pred = muti_results.predict(X_result,num_iteration=muti_results.best_iteration_)

print(y_pred)
DATA={
    'user_id': users,
    'predict': y_pred
}
dataframe = pd.DataFrame(DATA)

dataframe.to_csv("E:/BDCI Match/sample_submission.csv",sep=',',index = False,columns = ["user_id","predict"])
```
