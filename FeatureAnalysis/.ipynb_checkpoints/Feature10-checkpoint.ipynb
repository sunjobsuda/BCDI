```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import accuracy_score
from sklearn.feature_selection import VarianceThreshold
import warnings
from sklearn.preprocessing import OneHotEncoder
import zipfile
warnings.filterwarnings('ignore')
traindataFile = u"E:/BDCI Match/train.zip"


z = zipfile.ZipFile(traindataFile)
df = pd.read_csv(z.open(z.namelist()[0]))
# df = df[["service_type","current_service"]]
df.head(2)
```

```python
#非数值型变量object 转 float. 并补全缺失值
cols = df.columns[df.dtypes.eq('object')]
df[cols] = df[cols].apply(pd.to_numeric, errors='coerce')
```

```python
print(df.isnull().any())
```

```python
fee_null = pd.isnull(df['2_total_fee'])
data_null = df[fee_null == True]
data_null
```

```python
#求所有相似记录的2_total_fee的平均值
fee_2_mean_estimated = df.loc[(df['current_service']==89016252)&(df['1_total_fee']>110)&(df['1_total_fee']<130)]["2_total_fee"].mean()
print(fee_2_mean_estimated)
df.iloc[316212:316213,:]["2_total_fee"] =  fee_2_mean_estimated
df.iloc[316212:316213,:]
```

```python
#求所有相似记录的3_total_fee的平均值
fee_3_mean_estimated = df.loc[(df['current_service']==89016252)&(df['1_total_fee']>110)&(df['1_total_fee']<130)]["3_total_fee"].mean()
print(fee_2_mean_estimated)
df.iloc[316212:316213,:]["3_total_fee"] =  fee_3_mean_estimated
df.iloc[316212:316213,:]
```

```python
fee_null = pd.isnull(df['age'])
data_null = df[fee_null == True]
data_null
```

```python
temp = df[pd.isnull(df['gender']) == True]

for index, row in temp.iterrows():
    datas = df.loc[(df['current_service']==row["current_service"])]
    a = datas.sample(n=1)
    print(a["gender"].mean())
    df.iloc[index:index+1,:]["gender"] =  a["gender"].mean()
    print(df.iloc[index:index+1,:]["gender"])
```

```python
temp = df[pd.isnull(df['age']) == True]

for index, row in temp.iterrows():
    datas = df.loc[(df['current_service']==row["current_service"])]
    a = datas.sample(n=10)
    val = float(int(a["age"].mean()))
    print(val)
    df.iloc[index:index+1,:]["age"] =  val
    print(df.iloc[index:index+1,:]["age"])
```

```python
temp = df.loc[(df['gender']==0.0)]

for index, row in temp.iterrows():
    datas = df.loc[(df['current_service']==row["current_service"])&(df['gender']!=0)]
    a = datas.sample(n=10)
    val = a["gender"].mode().iloc[0]
    df.iloc[index:index+1,:]["gender"] = val
    print(df.iloc[index:index+1,:]["gender"])
```

```python
temp = df.loc[(df['age']==0.0)]

for index, row in temp.iterrows():
    datas = df.loc[(df['current_service']==row["current_service"])&(df['age']!=0)]
    a = datas.sample(n=10)
    val = float(int(a["age"].mean()))
    df.iloc[index:index+1,:]["age"] = val
    print(df.iloc[index:index+1,:]["age"])
```
